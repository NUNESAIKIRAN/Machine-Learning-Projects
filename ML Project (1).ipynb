### Identify the input and output/target variables

```python
X = df.drop(columns="quality")
```

```python
X.head()
```

```python
y = df['quality']
```

```python
y.head()
```

### Split Input and Outputs

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
X_train.head()
```

### Data preparation on Test Data

```python
X_train_num = X_train.drop(columns=["color","good"])
```

```python
X_train_num.shape
```

```python
X_train_cat = X_train[['color',"good"]]
```

```python
X_train_cat.shape
```

### Numerical Feature Transformation

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_num_trans = pd.DataFrame(scaler.fit_transform(X_train_num),columns=scaler.get_feature_names_out(),index=X_train_num.index)
```

```python
X_train_num_trans.head()
```

### Categorical Feature Transformation

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encode = OneHotEncoder(sparse_output=False,handle_unknown='ignore')
```

```python
X_train_cat_trans = pd.DataFrame(encode.fit_transform(X_train_cat),
                                 columns=encode.get_feature_names_out(),
                                 index = X_train_cat.index)
```

```python
X_train_cat_trans 
```

```python
X_test_num = X_test.drop(columns=['color',"good"])
```

```python
X_test_num.shape
```

```python
X_test_cat = X_test[['color',"good"]]
```

```python
X_test_cat.shape
```

###  Data Preparation On Test Data

```python
X_test_num_trans = pd.DataFrame(scaler.transform(X_test_num),columns=scaler.get_feature_names_out(),index = X_test_num.index)
```

```python
X_test_num_trans
```

```python
X_test_cat.head()
```

```python
X_test_cat_trans = pd.DataFrame(encode.transform(X_test_cat),columns=encode.get_feature_names_out(),index = X_test_cat.index)
```

```python
X_test_cat_trans
```

### Concatenation

```python
X_train_transformed = pd.concat([X_train_num_trans,X_train_cat_trans],axis=1)
```

```python
X_test_transformed = pd.concat([X_test_num_trans,X_test_cat_trans],axis = 1)
```

```python
#X_train_transformed,X_test_transformed
```

### Model Building

### 1.Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
regresion = LogisticRegression()
```

```python
%time regresion.fit(X_train_transformed,y_train)
```

```python
%%time 
y_test_pred = regresion.predict(X_test_transformed)
```

### Evalution

```python
from sklearn import metrics
```

```python
log_reg = metrics.accuracy_score(y_test_pred,y_test)
```

```python
log_reg
```

### 2.KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier()
```

```python
%time classifier.fit(X_train_transformed,y_train)
```

```python
%%time
y_pred1 = classifier.predict(X_test_transformed)
```

### Evalution

```python
Knn_acc = metrics.accuracy_score(y_pred1,y_test)
```

```python
 Knn_acc
```

### 3.Support Vector Machines

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
%time model.fit(X_train_transformed,y_train)
```

```python
%%time 
y_pred2 = model.predict(X_test_transformed)
```

### Evalution
